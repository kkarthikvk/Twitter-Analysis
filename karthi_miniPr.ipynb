```python
import re 
import tweepy 
from tweepy import OAuthHandler 
from textblob import TextBlob 
```

```python
#enter your own keys
consumer_key = '**************************'
consumer_secret = '***********************************'
access_token = '*****************************************************'
access_token_secret = '************************************'
  
```

```python
auth = OAuthHandler(consumer_key, consumer_secret) 
auth.set_access_token(access_token,access_token_secret)
api = tweepy.API(auth)
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()) 
```

```python
def get_tweet_sentiment(tweet): 
    analysis = TextBlob(clean_tweet(tweet)) 
    if analysis.sentiment.polarity > 0: 
        return 'positive'
    elif analysis.sentiment.polarity == 0: 
         return 'neutral'
    else: 
         return 'negative'
```

```python
search = input("Enter the hastag to be searched:")
```

```python
tweets = [] 
fetched_tweets = api.search(q = search, count = 100,lang = 'English') 
for tweet in fetched_tweets: 
    parsed_tweet = {} 
    parsed_tweet['text'] = tweet.text 
    parsed_tweet['sentiment'] = get_tweet_sentiment(tweet.text) 
    if tweet.retweet_count > 0: 
        if parsed_tweet not in tweets: 
            tweets.append(parsed_tweet) 
    else: 
        tweets.append(parsed_tweet) 
```

```python
ptweets = [tweet for tweet in tweets if tweet['sentiment'] == 'positive'] 
print("Positive tweets percentage: {} %".format(100*len(ptweets)/len(tweets))) 
ntweets = [tweet for tweet in tweets if tweet['sentiment'] == 'negative'] 
print("Negative tweets percentage: {} %".format(100*len(ntweets)/len(tweets))) 
print("Neutral tweets percentage: {} % ".format(100*(len(tweets) - len(ntweets) - len(ptweets))/len(tweets))) 
print("\n\nPositive tweets:") 
for tweet in ptweets[:10]: 
    print(tweet['text']) 
print("\n\nNegative tweets:") 
for tweet in ntweets[:10]: 
    print(tweet['text']) 
  
```

```python
df.drop('sentiment',axis=1,inplace=True)
```

```python
import re 
def rmno(text):
        tt = re.split('\W+',text)
        return tt

df['text'] = df['text'].apply(lambda x : rmno(x))
```

```python
import nltk
stpw = nltk.corpus.stopwords.words('english')
def rmstp(text):
    tt = [char for char in text if char not in stpw]
    return tt
df['text'] = df['text'].apply(lambda x:rmstp(x))
df.head(10)
```

```python

ps = nltk.PorterStemmer()
def rmstmer(text):
    tt = [ps.stem(char.lower()) for char in text]
    return tt
df['text'] = df['text'].apply(lambda x:rmstmer(x))
print(df.head())
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
comment_words = ' '
for i in df['text']: 
    for words in i:
        comment_words = comment_words + words + ' ' 
wordcloud = WordCloud(width = 600, height = 600, 
                background_color='white' , 

                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (5, 5), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0)
```
